```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import math

def calculate_figsize(num_ice, aspect_ratio=1.2):
    fig_width = 5  # 默认图形宽度
    fig_height = fig_width / aspect_ratio  # 默认图形高度

    num_rows = int(math.sqrt(num_ice))
    num_cols = math.ceil(num_ice / num_rows)

    figsize_width = fig_width * num_cols
    figsize_height = fig_height * num_rows

    return num_rows, num_cols, (figsize_width, figsize_height)

calculate_figsize(2)
```

```python
def get_mean_score_with_group(result_file):
    file_name = result_file.split('/')[-1][:-5]
    with open(result_file, 'r') as f:
        result_info = json.load(f)
    num_ice = len(result_info['0']['ice_idx'])
    idx = [int(x) for x in result_info.keys()]
    label = [info['label'] for _, info in result_info.items()]
    pred = [info['prediction'] for _, info in result_info.items()]
    ice_scores = [info['ice_score'] for _, info in result_info.items()]
    ice_scores = np.array(ice_scores)
    df_dict = {'idx': idx, 'label': label, 'pred': pred}
    for ith, ice_score in enumerate(ice_scores.T):
        df_dict[f'score_{ith}'] = ice_score
    df = pd.DataFrame(df_dict)
    df['flag'] = df['label'] == df['pred']
    df['flag'] = df['flag'].apply(lambda x: 'correct' if x else 'wrong')
    ddf_group = df.groupby(['flag'])
    ddf_mean = ddf_group.mean().T
    ddf_mean = ddf_mean.iloc[3:, :]
    return ddf_mean
```

```python
ice_num_list = [1, 2, 4, 8, 16, 32]
result_path = '/apdcephfs_cq8/share_2992827/public/yifeiyu/OpenICL/icl_inference_output_ascending/Topk'
model = 'gpt-neo-2.7B'
sbert = 'PMB2'
task = 'SST-5'
metric_file = f'{result_path}/{task}_results.txt'
df = pd.read_csv(metric_file, sep=',', header=None)
df.columns = ['task', 'ice_num', 'model', 'sbert', 'score']
df = df[(df['task'] == task) & (df['model'] == model) & (df['sbert'] == sbert)]

result_file_list = [f'{result_path}/{task}_{ice_num}_{model}_{sbert}.json' for ice_num in ice_num_list]

num_rows, num_cols, figsize = calculate_figsize(len(ice_num_list))

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=figsize)

for ith, (result_file, ice_num) in enumerate(zip(result_file_list, ice_num_list)):

    ax = axes[ith // num_cols, ith % num_cols]

    ddf_mean = get_mean_score_with_group(result_file)

    if ddf_mean.shape[0] <= 1:
        ddf_mean.plot(kind='hist', y=['correct', 'wrong'], color=['green', 'red'], ax=ax, alpha=0.5, density=True)
    else:
        ddf_mean.plot(kind='hist', y=['correct', 'wrong'], color=['green', 'red'], ax=ax, alpha=0.3, density=True)
        ddf_mean.plot(kind='kde', y=['correct', 'wrong'], color=['green', 'red'], ax=ax)
    
    acc = df['score'][df['ice_num'] == ice_num].values[0]

    ax.set_xlim(4, 7)
    ax.set_xlabel(f'Score_{ith}')
    ax.set_ylabel('Density')
    ax.set_title(f'Number of ICE: {ice_num}')
    ax.text(0.28, 0.10, f'Acc: {acc:.4f}', fontsize=12, transform=ax.transAxes, horizontalalignment='right', verticalalignment='top')
    ax.legend()
plt.suptitle(f'Distribution of mean ICE Scores of [{task}/{model}/{sbert}]')
plt.tight_layout()
plt.show()
```
